```python
import numpy as np
```

```python
inputs=[[-2,-1],[25,6],[17,4],[-15,-6]]
```

```python
target=[1,0,0,1]
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
w1= 0.5
w2 = 0.7
w3= -0.6
w4= 0.9
w5= 0.4
w6= 0.8
b1= 2
b2= 1.2
b3= 4.5
```

```python
error_tracking=[]
```

```python
for epoch in range (1000):
  i=0
  for weight,height in inputs:
    h1= sigmoid(w1*weight+w2*height+b2)
    h2= sigmoid(w3*weight+w4*height+b1)
    o1= sigmoid(w5*h1+w6*h2+b3)

#Backward Propagation for output neuron

    dl_by_dw5= o1*(1-o1)*h1*(o1-target[i])
    dl_by_dw6= o1*(1-o1)*h2*(o1-target[i])
    dl_by_b3 =o1*(1-o1)*(o1-target[i])

#Backward Propagation for hidden neuron 1
    dl_by_dw1= h1*(1-h1)*o1*(1-o1)*(o1-target[i])*weight
    dl_by_dw2= h1*(1-h1)*o1*(1-o1)*(o1-target[i])*height
    dl_by_b2 = h1*(1-h1)*o1*(1-o1)*(o1-target[i])


#Backward Propagation for hidden neuron 2
    dl_by_dw3= h2*(1-h2)*o1*(1-o1)*(o1-target[i])*weight
    dl_by_dw4= h2*(1-h2)*o1*(1-o1)*(o1-target[i])*height
    dl_by_b1 = h2*(1-h2)*o1*(1-o1)*(o1-target[i])

    lr=0.01 #Learning Rate
    w5=w5-lr*dl_by_dw5
    w6=w6-lr*dl_by_dw6
    b3=b3-lr*dl_by_b3
    w1=w1-lr*dl_by_dw1
    w2=w2-lr*dl_by_dw2
    b2=b2-lr*dl_by_b2
    w3=w3-lr*dl_by_dw3
    w4=w4-lr*dl_by_dw4
    b1=b1-lr*dl_by_b1

    error=0.5*(target[i]-o1)**2

    i=+1
  error_tracking.append(error)
```
