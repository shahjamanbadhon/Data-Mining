```python
import numpy as np
import pandas as pd
```

```python
inputs = [[-2, -1],[25, 6],[17, 4],[-15, -6]]
```

```python
target = [1,0,0,1]
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
w1 = 0.5
w2 = 0.7
w3 = -0.6
w4 = 0.9
w5 = 0.4
w6 = 0.8
b1 = 2
b2 = 1.2
b3 = 4.5
learning_rate = 0.01
error_tracking = []
```

```python
for epoch in range(1000):
  i = 0
  error = 0
  for weight, height in inputs:

    # Calculate the output of first neuron
    h1 = sigmoid(weight*w1 + height*w2 + b2)
    # Calculate the output of second neuron
    h2 = sigmoid(weight*w3 + height*w4 + b1)
    # Calculate the output of final layer neuron
    o1 = sigmoid(h1*w5 + h2*w6 + b3)

    # Backward propagation for output neuron
    dl_by_dw5 = o1*(1-o1)*h1*(o1 - target[i])
    dl_by_dw6 = o1*(1-o1)*h2*(o1 - target[i])
    dl_by_db3 = o1*(1-o1)*(o1 - target[i])

    # Backward propagatoion of hidden neuron 1
    dl_by_dw1 = (h1*(1-h1)*o1*(1-o1)*(o1-target[i]*weight))
    dl_by_dw2 = (h1*(1-h1)*o1*(1-o1)*(o1-target[i]*height))
    dl_by_db2 = (h1*(1-h1)*o1*(1-o1)*(o1-target[i]))

    # Backward propagation of hidden neuron 2
    dl_by_dw3 = (h2*(1-h2)*o1*(1-o1)*(o1-target[i]*weight))
    dl_by_dw4 = (h2*(1-h2)*o1*(1-o1)*(o1-target[i]*height))
    dl_by_db1 = (h2*(1-h2)*o1*(1-o1)*(o1-target[i]))

    # Update the weight value
    w1 = w1 - learning_rate*dl_by_dw1
    w2 = w2 - learning_rate*dl_by_dw2
    w3 = w3 - learning_rate*dl_by_dw3
    w4 = w4 - learning_rate*dl_by_dw4
    w5 = w5 - learning_rate*dl_by_dw5
    w6 = w6 - learning_rate*dl_by_dw6

    # Update the bias value
    b1 = b1 - learning_rate*dl_by_db1
    b2 = b2 - learning_rate*dl_by_db2
    b3 = b3 - learning_rate*dl_by_db3
    #error=0.5*(target[i]-o1)**2
    error = error + (0.5 * (o1 - target[i]))

    i = i + 1
  error_tracking.append(error/4)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1000), error_tracking)
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.show()
```
