```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
inputs = [[10, 20, 30],[50, 10, 20],[60, 60, 5]]
```

```python
target = [1,0,1]
```

```python
w1 = 0.2
w2 = 0.5
w3 = 0.9
w4 = 0.7
w5 = -0.8
w6 = 0.2
w7 = 0.5
w8 = 0.7
w9 = -0.8
w10 = 0.3
w11 = 0.2
w12 = 0.1
b1 = 2
b2 = 1.2
b3 = 4.5
b4 = 2.2
learning_rate = 0.01
error_tracking = []
```

```python
for epoch in range(2000):
  i = 0
  error = 0
  for x1, x2, x3 in inputs:

    # Calculate the output of first neuron
    h1 = sigmoid(x1*w1 + x2*w2 + x3*w3 + b1)
    # Calculate the output of second neuron
    h2 = sigmoid(x1*w4 + x2*w5 + x3*w6 + b2)
    # Calculate the output of third neuron
    h3 = sigmoid(x1*w7 + x2*w8 + x3*w9 + b3)
    # Calculate the output of final layer neuron
    o1 = sigmoid(h1*w10 + h2*w11 + h3*w12 + b4)

    # Backward propagation for output neuron
    dl_by_dw10 = o1*(1-o1)*h1*(o1 - target[i])
    dl_by_dw11 = o1*(1-o1)*h2*(o1 - target[i])
    dl_by_dw12 = o1*(1-o1)*h2*(o1 - target[i])
    dl_by_db4 = o1*(1-o1)*(o1 - target[i])


    # Backward propagatoion of hidden neuron 1
    dl_by_dw1 = (h1*(1-h1)*o1*(1-o1)*(o1-target[i]*x1))
    dl_by_dw2 = (h1*(1-h1)*o1*(1-o1)*(o1-target[i]*x2))
    dl_by_dw3 = (h1*(1-h1)*o1*(1-o1)*(o1-target[i]*x3))
    dl_by_db1 = (h1*(1-h1)*o1*(1-o1)*(o1-target[i]))


    # Backward propagation of hidden neuron 2
    dl_by_dw4 = (h2*(1-h2)*o1*(1-o1)*(o1-target[i]*x1))
    dl_by_dw5 = (h2*(1-h2)*o1*(1-o1)*(o1-target[i]*x2))
    dl_by_dw6 = (h2*(1-h2)*o1*(1-o1)*(o1-target[i]*x3))
    dl_by_db2 = (h2*(1-h2)*o1*(1-o1)*(o1-target[i]))


    # Backward propagation of hidden neuron 2
    dl_by_dw7 = (h3*(1-h3)*o1*(1-o1)*(o1-target[i]*x1))
    dl_by_dw8 = (h3*(1-h3)*o1*(1-o1)*(o1-target[i]*x2))
    dl_by_dw9 = (h3*(1-h3)*o1*(1-o1)*(o1-target[i]*x3))
    dl_by_db3 = (h3*(1-h3)*o1*(1-o1)*(o1-target[i]))

    # Update the weight value
    w1 = w1 - learning_rate*dl_by_dw1
    w2 = w2 - learning_rate*dl_by_dw2
    w3 = w3 - learning_rate*dl_by_dw3
    w4 = w4 - learning_rate*dl_by_dw4
    w5 = w5 - learning_rate*dl_by_dw5
    w6 = w6 - learning_rate*dl_by_dw6
    w7 = w7 - learning_rate*dl_by_dw7
    w8 = w8 - learning_rate*dl_by_dw8
    w9 = w9 - learning_rate*dl_by_dw9
    w10 = w10 - learning_rate*dl_by_dw10
    w11 = w11 - learning_rate*dl_by_dw11
    w12 = w12 - learning_rate*dl_by_dw12

    # Update the bias value
    b1 = b1 - learning_rate*dl_by_db1
    b2 = b2 - learning_rate*dl_by_db2
    b3 = b3 - learning_rate*dl_by_db3
    b4 = b4 - learning_rate*dl_by_db4
    error = error + (0.5 * (o1 - target[i]))

    i = i + 1
  error_tracking.append(error/4)
```

```python
plt.plot(range(2000), error_tracking)
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.show()
```
