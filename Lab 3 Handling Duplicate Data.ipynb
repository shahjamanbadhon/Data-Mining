```python
data = {
    'id': [1, 2, 3, 4, 5, 6],
    'age': [25, 27, 29, 31, 120, 28],   # 120 is an outlier
    'salary': [3000, 3200, 3500, 4000, 5000, 100000]  # 100000 is an outlier
}
```

```python
df = pd.DataFrame(data)
```

```python
print("Original DataFrame:")
print(df)
```

```python
outlier_info = {}

for i in ["age", "salary"]:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    x = (df[i] < lower_bound) | (df[i] > upper_bound)
    outlier_info[i] = df[x]
    

outlier_info
```
